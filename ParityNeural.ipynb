```python
import dataGenerator
import NeuralNets as NN
import pandas as pd
import numpy as np
```

```python
dg = dataGenerator.ParityDataGenerator()
df = dg.generate_data()
print(df)
```

```python
def get_base_model_obj():
    input = NN.InputLayer(n_neurons=4)
    hid1 = NN.DenseLayer(n_neurons=4, previous_layer=input)
    # hid1 = NN.DenseLayer(n_neurons=4, previous_layer=hid1)
    hid2 = NN.DenseLayer(n_neurons=1, previous_layer=hid1, isOutput=True)
    mdl = NN.Model(input_layer=input, output_layer=hid2)
    return mdl
```

```python
epoch_num = []
logs = []
for lr in np.arange(0.05, 0.5, 0.05):
    mdl = get_base_model_obj()
    log = get_base_model_obj().fit(df.drop(columns=['label']).values, df['label'].values.reshape(-1,1), lr=lr, momentum=0,max_epochs=100)
    logs.append(log)
    epoch_num.append(len(log))
```

```python
epoch_num = []
for i in range(len(logs)):
    epoch_num.append(len(logs[i]))
print(epoch_num)
```

```python
from keras.layers import Dense, Input
from keras.models import Model
from keras import optimizers
from keras import regularizers
import os
```

```python
df = df.sample(frac=1).astype(np.float64)

input = Input([4,])
d = Dense(4,activation='sigmoid',activity_regularizer=regularizers.l1(0.01))(input)
o = Dense(1,activation='sigmoid',activity_regularizer=regularizers.l1(0.01))(d)
optimizer = optimizers.Adam(lr=1e-2)
model = Model(inputs=input,outputs=o)
model.compile(optimizer, loss='mse',metrics=['acc'])
model.fit(df.drop(columns=['label']).values, df['label'].values,epochs=10000,batch_size=1,verbose=0)
```

```python
print(model.predict(df.drop(columns=['label'])))
print(df['label'])
print(model.evaluate(df.drop(columns=['label']).values, df['label'].values))
```

```python
from keras.layers import Dense, Input
from keras.models import Model
from keras import optimizers
import os
# try different dimensionality
result = []
for i in range(1,7,1):
    dg = dataGenerator.ParityDataGenerator(i)
    df = dg.generate_data()
    df = df.sample(frac=1).astype(np.float64)

    input = Input([i,])
    d = Dense(4,activation='sigmoid')(input)
    o = Dense(1,activation='sigmoid')(d)
    optimizer = optimizers.Adam(lr=5e-2)
    model = Model(inputs=input,outputs=o)
    model.compile(optimizer, loss='mse',metrics=['acc'])
    model.fit(df.drop(columns=['label']).values, df['label'].values,epochs=100,batch_size=1,verbose=0)
    result.append(model.evaluate(df.drop(columns=['label']).values, df['label'].values))
```
